```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import numpy as np

# Charger le modèle YOLOv8
model = YOLO("runs/detect/train6/weights/best.pt")

# Charger l'image
image_path = r"C:\Users\emnad\Downloads\datasetsplit1\test\Bulletin_de_soin\0747--9550703--20230705_page_0.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Conversion pour Matplotlib

# Détection avec un seuil de confiance bas
results = model(image_path, conf=0.05)

# Paramètres personnalisés pour les boîtes
box_alpha = 0.2  # Transparence du remplissage
line_width = 1    # Épaisseur des contours
font_scale = 0.6  # Taille du texte
text_color = (255, 0, 0)  # Couleur du texte (bleu)
box_color = (0, 255, 0)   # Couleur des boîtes (vert)

# Créer une copie pour l'annotation
annotated_image = image.copy()

for result in results:
    for box in result.boxes:
        # Coordonnées de la boîte
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        
        # Dessiner le rectangle avec transparence
        overlay = annotated_image.copy()
        cv2.rectangle(overlay, (x1, y1), (x2, y2), box_color, -1)  # Remplissage
        annotated_image = cv2.addWeighted(overlay, box_alpha, annotated_image, 1 - box_alpha, 0)
        
        # Contour de la boîte
        cv2.rectangle(annotated_image, (x1, y1), (x2, y2), box_color, line_width)
        
        # Texte (classe + confiance)
        label = f"{result.names[int(box.cls)]} {box.conf[0]:.2f}"
        cv2.putText(annotated_image, label, (x1, y1 - 5),
                   cv2.FONT_HERSHEY_SIMPLEX, font_scale, text_color, 1)

# Affichage
plt.figure(figsize=(12, 10))
plt.imshow(annotated_image)
plt.axis("off")

# Enregistrer le résultat si besoin
output_path = "detection_result.jpg"
cv2.imwrite(output_path, cv2.cvtColor(annotated_image, cv2.COLOR_RGB2BGR))
print(f"Résultat sauvegardé sous : {output_path}")

plt.show()
```

```python
import pandas as pd
import re
import os

# Chemins importants
input_file = "annotationtxt.txt"  # Votre fichier texte original
output_csv = "annotations_formatted.csv"
new_image_dir = r"C:\Users\emnad\Downloads\data_dir\images"

# Lire le fichier d'origine
with open(input_file, 'r', encoding='utf-8') as f:
    lines = [line.strip() for line in f if line.strip()]

# Traitement des données
data = []
for line in lines:
    # Séparation du chemin et du texte
    parts = re.split(r'\t', line)  # Utilisation de la tabulation comme séparateur
    if len(parts) == 2:
        old_path = parts[0].strip('"')
        text = parts[1]
        
        # Extraire le nom du fichier
        filename = os.path.basename(old_path)
        
        # Créer le nouveau chemin complet
        new_path = os.path.join(new_image_dir, filename)
        
        data.append({
            "Image Path": new_path.replace("\\", "/"),  # Normalisation des chemins
            "Text": text
        })

# Création du DataFrame
df = pd.DataFrame(data)

# Vérification des chemins
print("Vérification des premiers chemins:")
print(df.head(3))

# Sauvegarde en CSV
df.to_csv(output_csv, index=False, encoding='utf-8-sig')

print(f"\nFichier CSV généré avec succès: {output_csv}")
print(f"Nombre total d'entrées: {len(df)}")
```

```python
from PIL import Image
import numpy as np

# Charger l'image
image = Image.open(image_path).convert("RGB")
image_np = np.array(image)  # Convertir en tableau NumPy pour faciliter le cropping
```

```python
# Dossier pour sauvegarder les crops
import os
output_dir = "crops"
os.makedirs(output_dir, exist_ok=True)

# Cropper et sauvegarder chaque zone
for i, box in enumerate(boxes):
    x_min, y_min, x_max, y_max = box.tolist()  # Convertir en liste
    x_min, y_min, x_max, y_max = int(x_min), int(y_min), int(x_max), int(y_max)

    # Cropper la région
    cropped_image = image_np[y_min:y_max, x_min:x_max]
    cropped_image = Image.fromarray(cropped_image)

    # Sauvegarder l'image croppée
    crop_path = os.path.join(output_dir, f"crop_{i:03d}.png")
    cropped_image.save(crop_path)
    print(f"Sauvegardé : {crop_path}")
```

```python
import easyocr
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger l'image
image_path = r"C:\Users\emnad\Downloads\crops\crop_009.png"
#image_path = r"C:\Users\emnad\Downloads\dataset_split\train\Bulletin_de_soin\9905--8309317--20231031_page_0.jpg"
image = cv2.imread(image_path)

# Convertir en niveaux de gris
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Normalisation et binarisation
normalized = cv2.normalize(gray, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
blurred = cv2.GaussianBlur(normalized, (5, 5), 0)
binary = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 31, 5)

# Sauvegarde de l'image prétraitée
preprocessed_image_path = "preprocessed_image1.jpg"
cv2.imwrite(preprocessed_image_path, binary)

# OCR avec EasyOCR
reader = easyocr.Reader(['fr'], gpu=False)
results = reader.readtext(image_path)

# Trier les résultats par position verticale, puis horizontale
sorted_results = sorted(results, key=lambda x: (x[0][0][1], x[0][0][0]))

# Affichage structuré
print("\n📄 Texte extrait organisé :\n")
for (bbox, text, prob) in sorted_results:
    print(f"{text} (Probabilité: {prob:.2f})")
```
