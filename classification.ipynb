```python
import cv2
import pytesseract
import numpy as np
import matplotlib.pyplot as plt

# D√©finir le chemin vers Tesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

# Charger l'image
image_path = r"C:\Users\emnad\Downloads\bs2.jpg"
image = cv2.imread(image_path)

if image is None:
    print("‚ùå Impossible de charger l'image.")
else:
    # Convertir en niveaux de gris
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Am√©liorer le contraste (CLAHE)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    gray = clahe.apply(gray)

    # Binarisation
    _, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # D√©tection du texte avec position
    data = pytesseract.image_to_data(binary, lang="fra+ara", output_type=pytesseract.Output.DICT)

    # Extraire les positions et textes valides
    text_lines = []
    for i in range(len(data["text"])):
        text = data["text"][i].strip()
        if text:
            x, y, w, h = data["left"][i], data["top"][i], data["width"][i], data["height"][i]
            text_lines.append((y, x, w, h, text))  # Stocker Y pour trier

    # Trier les lignes par position verticale
    text_lines.sort()

    # S√©parer les phrases Fran√ßais / Arabe
    french_texts = []
    arabic_texts = []

    for _, x, _, _, text in text_lines:
        if x < image.shape[1] // 2:  # Texte √† gauche (Fran√ßais)
            french_texts.append(text)
        else:  # Texte √† droite (Arabe)
            arabic_texts.append(text)

    # Associer les paires
    paired_texts = list(zip(french_texts, arabic_texts))

    # Afficher les paires bien align√©es
    print("\nüìú Paires Fran√ßais - Arabe :\n")
    for fr, ar in paired_texts:
        print(f"üá´üá∑ {fr}  ---  üá¶üá™ {ar}")

    # Afficher l'image avec d√©tection des zones
    image_with_boxes = image.copy()
    for y, x, w, h, text in text_lines:
        cv2.rectangle(image_with_boxes, (x, y), (x + w, y + h), (0, 255, 0), 2)

    plt.figure(figsize=(10, 10))
    plt.imshow(cv2.cvtColor(image_with_boxes, cv2.COLOR_BGR2RGB))
    plt.title("Zones d√©tect√©es")
    plt.axis('off')
    plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

IMG_SIZE = 224  # Taille des images pour MobileNetV2
BATCH_SIZE = 16  # Taille du batch
EPOCHS = 10  # Nombre d'it√©rations d'entra√Ænement
DATASET_PATH = "C:/Users/emnad/Downloads/dataclass"  # Dossier contenant les images

# üîπ G√©n√©ration des donn√©es avec augmentation pour √©viter le sur-apprentissage
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Normalisation des pixels entre 0 et 1
    validation_split=0.2,  # 20% des images seront utilis√©es pour la validation
    rotation_range=20,  
    width_shift_range=0.2, 
    height_shift_range=0.2,
    shear_range=0.2, 
    zoom_range=0.2,
    horizontal_flip=True
)

# üîπ Chargement des images depuis le dossier
train_generator = train_datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',  # Classification binaire : Bulletin de soin vs Ordonnance
    subset='training'
)

val_generator = train_datagen.flow_from_directory(
    DATASET_PATH,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)

# üîπ Chargement du mod√®le MobileNetV2 pr√©-entra√Æn√©
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))

# üîπ On "g√®le" les couches du mod√®le pour ne pas les modifier
base_model.trainable = False

# üîπ Ajout de nouvelles couches pour la classification
x = base_model.output
x = GlobalAveragePooling2D()(x)  # Permet de r√©duire la dimension
x = Dense(128, activation='relu')(x)  # Couche interm√©diaire
x = Dense(64, activation='relu')(x)  # Une autre couche
predictions = Dense(1, activation='sigmoid')(x)  # 1 Neurone pour classification binaire

# üîπ Cr√©ation du mod√®le final
model = Model(inputs=base_model.input, outputs=predictions)

# üîπ Compilation du mod√®le
model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

# üîπ Entra√Ænement du mod√®le
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS
)

# üîπ Sauvegarde du mod√®le
model.save("modele_classification_bs_vs_ordonnance.h5")

# üîπ Affichage des courbes de pr√©cision et de perte
plt.figure(figsize=(12, 4))

# Courbe de pr√©cision
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label="Entra√Ænement")
plt.plot(history.history['val_accuracy'], label="Validation")
plt.title("Pr√©cision du mod√®le")
plt.legend()

# Courbe de perte
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label="Entra√Ænement")
plt.plot(history.history['val_loss'], label="Validation")
plt.title("Perte du mod√®le")
plt.legend()

plt.show()
```

```python
import os
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, Flatten, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
```

```python
# D√©finition des chemins
DATASET_DIR = "C:/Users/emnad/Downloads/dataPI"
CLASSES = ["bulletin_de_soin", "ordonnance"]
IMG_SIZE = 224  # Taille standard pour MobileNetV2

# Chargement des images et labels
X, y = [], []
for label, category in enumerate(CLASSES):
    folder_path = os.path.join(DATASET_DIR, category)
    for img_name in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_name)
        img = cv2.imread(img_path)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        img = img / 255.0  # Normalisation
        X.append(img)
        y.append(label)

# Conversion en tableau numpy
X = np.array(X)
y = np.array(y)

# S√©paration en train / test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# Chargement du mod√®le pr√©-entra√Æn√©
base_model = MobileNetV2(weights="imagenet", include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3))
base_model.trainable = False  # On freeze les poids

# Ajout des couches personnalis√©es
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation="relu")(x)
x = Dropout(0.3)(x)
output = Dense(1, activation="sigmoid")(x)  # 1 neurone avec sigmoid pour binaire

model = Model(inputs=base_model.input, outputs=output)

# Compilation du mod√®le
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
# G√©n√©ration de donn√©es augment√©es
datagen = ImageDataGenerator(rotation_range=15, zoom_range=0.2, horizontal_flip=True)

# Entra√Ænement
history = model.fit(datagen.flow(X_train, y_train, batch_size=32),
                    validation_data=(X_test, y_test),
                    epochs=10)
```

```python
# √âvaluation sur le test set
loss, acc = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {acc * 100:.2f}%")
```
